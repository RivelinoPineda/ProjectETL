### Proyecto ETL

```python
# Importacion de librerias
import pandas as pd
import numpy as np
import os
```

```python
# Definicion de paths
path_ = os.getcwd() #método para obtener el directorio de trabajo actual
dataset_name = 'jio_mart_items.csv' #Datos de lectura 
```

```python
# Carga de datos rawdata products
products = pd.read_csv(os.path.join(path_, dataset_name))
products
```

```python
# Create table category
list_category = products['category'].unique() #obtener los diferentes valores de las categorias
size_category = len(list_category)
list_id_category = list(range(1,size_category+1))#Hacer la lista
category = pd.DataFrame({'category_name': list_category,
                        'category_id': list_id_category})
category
```

```python
# Create table sub-category
list_sub_category = products['sub_category'].unique()
size_sub_category = len(list_sub_category)
list_id_sub_category = list(range(1,size_sub_category+1))
sub_category = pd.DataFrame({'sub_category_name': list_sub_category,
                            'sub_category_id': list_id_sub_category})
```

```python
# Trabajar en el update para manejar los productos que se repiten
products_value_counts = products['items'].value_counts().reset_index()#conteo de valores e indexar
products_value_counts = products_value_counts.rename(columns={'items': 'count', 'index': 'items'}) #renoembrar columnas 
products = pd.merge(products, products_value_counts, on='items', how='left')#products=left,products_va=right, on=indixe comun,
#how=uso de claves
products = products.query('count == 1.0')
```
